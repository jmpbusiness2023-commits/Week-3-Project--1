#### Hypothesis 1 : Movies with lower production budgets can still achieve high IMDb ratings

```python
low_budget_threshold = df_cleaned["production_budget"].quantile(0.25)  # bottom 25%
high_rating_threshold = df_cleaned["imdb_rating"].quantile(0.75)       # top 25%

low_budget_high_rating = df_cleaned[
    (df_cleaned["production_budget"] <= low_budget_threshold) &
    (df_cleaned["imdb_rating"] >= high_rating_threshold)
]

low_budget_high_rating[["title", "production_budget", "imdb_rating", "genre"]].sort_values(
    "imdb_rating", ascending=False
).head(10)
```

#### Hypothesis 2 : Are Hight IMDB Rated Movies More Profitable?

```python
df_cleaned.head()
```

```python
high_profit_threshold = df_cleaned["profit"].quantile(0.75)  # top 25%
high_rating_threshold = df_cleaned["imdb_rating"].quantile(0.75) # top 25%

high_profit_high_rating = df_cleaned[
    (df_cleaned["profit"] >= high_profit_threshold) &
    (df_cleaned["imdb_rating"] >= high_rating_threshold)
]

high_profit_high_rating[["title", "production_budget", "imdb_rating", "genre"]].sort_values(
    "imdb_rating", ascending=False
).head(10)
```
